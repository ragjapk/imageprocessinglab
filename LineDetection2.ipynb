# Perform line detection using finite difference kernels

Gradient of the given image $g(m,n)$ can be obtained using any of the gradient operators (Roberts, Sobel, Prewitt, etc) to get the magnitude the orientation:
$$M(m,n)=\sqrt{g_m^2(m,n)+g_n^2(m,n)}$$
and
$$\theta(m,n)=tan^{-1}[g_n(m,n)/g_m(m,n)$$

# Algorithm

\begin{enumerate}
\item Read an Image
\item Form the Sobel, Prewitt, Robert Kernels - discrete form
\item Convolve the image with the edge detection kernels and find out the gradients
\item Repeat the same for noised images.
\item Save the images
\end{enumerate}
# Code

```python
import numpy as np
from PIL import Image
from scipy import signal
import matplotlib.pyplot as plt
%matplotlib inline
img=Image.open('gaussian_noise.png').convert('L')
img=np.asarray(img)

#vertical kernel
fx=[[-1,0,1],[-2,0,2],[-1,0,1]]

#horizontal kernel
fy=[[-1,-2,-1],[0,0,0],[1,2,1]]

vertical=signal.convolve2d(img,fy)
horizontal=signal.convolve2d(img,fx)

plt.imshow(Image.fromarray(img),cmap='gray')
plt.figure()
plt.imshow(Image.fromarray(vertical),cmap='gray')
plt.figure()
plt.imshow(Image.fromarray(horizontal),cmap='gray')
```
